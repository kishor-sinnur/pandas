```python
new_coffee_data=pd.read_csv("./data/newadd.csv")
new_coffee_data["luxury"]=new_coffee_data["Coffee Type"].apply(lambda x:"primium" if x=="Espresso" else "ultra-luxury")
print(new_coffee_data)
```

```python
def determine_status(row):
    if row['price'] > 40 and row['Revenue'] > 900:
        return "limited edition"
    else:
        return "oversold"
new_coffee_data["status"]=new_coffee_data.apply(determine_status, axis=1)
new_coffee_data.head()
```

```python
new_coffee_data.groupby(["luxury"]).all()
```

```python
pd.merge(new_coffee_data,coffee, on="Coffee Type", how="inner") # same for inner join and outer join left join ,right join
```
